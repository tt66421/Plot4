```python
# インポート
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# プロットするデータ
x1 = np.arange(0, 1.01, 0.1)
x2 = [x**2 for x in x1]
xRoot = [x**0.5 for x in x1]
tanh = [np.tanh(10*x) for x in x1]
```

```python
# 図を出力

# yのデータのリスト
ylist = [x1, x2, xRoot, tanh]
# マーカーの形のリスト
markers = ['x', '+', 'o', 's']
# マーカーの色のリスト
colors = ['g', 'b', 'm', 'r']
# マーカーを繋ぐ線の種類のリスト
linestyles = ['dashed', 'dotted', 'solid', 'dashdot']
# 凡例のリスト
legends = ['$y=x$', '$y=x^2$', '$y=\sqrt{x}$', '$y=\\tanh(10x)$']

for i in range(len(ylist)):
    plt.plot(
        # x, y のデータ
        x1, ylist[i], 
        # マーカーの形
        markers[i],
        # マーカーサイズ
        markersize=10,
        # マーカーの枠の太さ
        markeredgewidth=3, 
        # マーカーの枠の色
        markeredgecolor=colors[i],  
        # マーカーの中身の色（マーカーが重なる場合は中抜きがよい）
        markerfacecolor='none',
        # マーカーを繋ぐ線の種類
        linestyle=linestyles[i], 
        # マーカーを繋ぐ線の色
        color = colors[i]
    )
    
# x,y 軸の範囲（少し広めにすると見易い）
plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)

# 共通のフォントサイズ
plt.rcParams["font.size"] = 14
# 軸目盛のフォントサイズ
plt.tick_params(labelsize=14)
# x軸ラベル
plt.xlabel('$x$', fontsize=18)
# y 軸ラベル
plt.ylabel('$y$', fontsize=18)
# 凡例の表示, マーカーの数, 位置, フォントサイズ
plt.legend(legends, numpoints=1, loc='best', fontsize=12)
# グラフのタイトル
plt.title('Plot for 4 data')

# 図の保存
# Supported formats: eps, jpeg, jpg, pdf, pgf, png, ps, raw, rgba, svg, svgz, tif, tiff.
plt.savefig('plot4.png', bbox_inches="tight")
plt.savefig('plot4.svg', bbox_inches="tight")
plt.savefig('plot4.pdf', bbox_inches="tight")
plt.savefig('plot4.eps', bbox_inches="tight")

# 図をリセット
plt.figure()
```
